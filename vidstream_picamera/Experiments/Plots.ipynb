```python
from matplotlib import pyplot as plt
import numpy as np

def extract_data(fname):
    with open(fname) as file:
        data_fps = []
        data_img_len = []
        for idx, line in enumerate(file):
            if idx > 0:
                elts = line.split(",")
                data_fps.append( float(elts[0]) / float(elts[1]) )
                data_img_len.append(float(elts[2]))
                #data_img_len.append( float(elts[2]) )
    return data_fps, data_img_len
```

# Capture continuous *versus* Capture sequence

```python
fname1 = "stream_perf_01.txt"
fname2 = "stream_perf_02.txt"
data_fps1 = extract_data(fname1)[0]
avg_fps1 = np.mean( data_fps1 )
data_fps2 = extract_data(fname2)[0]
avg_fps2 = np.mean( data_fps2 )
print( " seq: {} | cont: {}".format(avg_fps1, avg_fps2) )
plt.plot(np.arange(0, len(data_fps1), 1), data_fps1)
plt.plot(np.arange(0, len(data_fps2), 1), data_fps2)
plt.title("Capture type")
plt.xlabel("records", fontsize=18)
plt.ylabel("fps", fontsize=18)
plt.legend(['seq [avg fps={:.1f}]'.format(avg_fps1), 'cont [avg fps={:.1f}]'.format(avg_fps2)])
plt.ylim([0,30])
plt.show()
```

# Image format: "jpeg" *vs* "yuv" *vs* "rgb"
## Frame per seconds

```python
#Case 1: use JPEG
fname1 = "stream_perf_01.txt"
fname2 = "stream_perf_04.txt"
fname3 = "stream_perf_05.txt"
data_fps1 = extract_data(fname1)[0]
avg_fps1 = np.mean( data_fps1 )
data_fps2 = extract_data(fname2)[0]
avg_fps2 = np.mean( data_fps2 )
data_fps3 = extract_data(fname3)[0]
avg_fps3 = np.mean( data_fps3 )
print( " jpeg: {} | yuv: {} | rgb: {}".format(avg_fps1, avg_fps2,  avg_fps3) )
plt.plot(np.arange(0, len(data_fps1), 1), data_fps1)
plt.plot(np.arange(0, len(data_fps2), 1), data_fps2)
plt.plot(np.arange(0, len(data_fps3), 1), data_fps3)
plt.title("Image format")
plt.xlabel("records", fontsize=18)
plt.ylabel("fps", fontsize=18)
plt.legend(['jpeg [avg fps={:.1f}]'.format(avg_fps1), 'yuv [avg fps={:.1f}]'.format(avg_fps2), 'rgb [avg fps={:.1f}]'.format(avg_fps3)])
plt.ylim([0,30])
plt.show()
```
